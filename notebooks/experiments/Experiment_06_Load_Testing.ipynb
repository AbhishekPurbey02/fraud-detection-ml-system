```python
import requests
import time
import pandas as pd
```

```python
url = "http://127.0.0.1:5000/predict"
```

```python
sample_data = {
    "features": [
         0.0,
  -1.3598071336738,
  -0.0727811733098497,
  2.53634673796914,
  1.37815522427443,
  -0.338320769942518,
  0.462387777762292,
  0.239598554061257,
  0.0986979012610507,
  0.363786969611213,
  0.0907941719789316,
  -0.551599533260813,
  -0.617800855762348,
  -0.991389847235408,
  -0.311169353699879,
  1.46817697209427,
  -0.470400525259478,
  0.207971241929242,
  0.0257905801985591,
  0.403992960255733,
  0.251412098239705,
  -0.018306777944153,
  0.277837575558899,
  -0.110473910188767,
  0.0669280749146731,
  0.128539358273528,
  -0.189114843888824,
  0.133558376740387,
  -0.0210530534538215,
  149.62,
    ]
}
```

```python
num_requests = 500

success = 0
failed = 0

start_time = time.perf_counter()

for i in range(num_requests):

    try:
        response = requests.post(
            url,
            json=sample_data
        )

        if response.status_code == 200:
            success += 1
        else:
            failed += 1

    except:
        failed += 1

end_time = time.perf_counter()

total_time = end_time - start_time

throughput = success / total_time

print(f"Success: {success}")
print(f"Failed: {failed}")
print(f"Time: {total_time:.2f}")
print(f"Throughput: {throughput:.2f} req/sec")
```

```python
results = pd.DataFrame({
    "Metric":[
        "Total Requests",
        "Successful Requests",
        "Failed Requests",
        "Execution Time (sec)",
        "Throughput (req/sec)"
    ],
    "Value":[
        num_requests,
        success,
        failed,
        round(total_time,2),
        round(throughput,2)
    ]
})

results
```
