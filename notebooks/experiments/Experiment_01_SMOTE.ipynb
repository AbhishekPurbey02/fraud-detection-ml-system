```python
from sklearn.model_selection import train_test_split

# 1. Load the dataset (using our bulletproof path)
notebook_dir = os.getcwd()
data_path = os.path.abspath(os.path.join(notebook_dir, "..", "..", "data", "creditcard.csv"))
dataset = pd.read_csv(data_path)

# 2. Separate features (X) and target label (y)
# Assuming your target column is named 'Class' (0 = Normal, 1 = Fraud)
X = dataset.drop(columns=['Class'])
y = dataset['Class']

# 3. Perform the train-test split (e.g., 80% train, 20% test)
# stratify=y is critical here to keep the fraud percentage identical in both splits!
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y
)

# 4. Get the exact breakdown of the Training Dataset
train_counts = y_train.value_counts()
total_train = len(y_train)

print("========== NORMAL TRAINING DATASET PROFILE ==========")
print(f"Total Samples in Training Set: {total_train}")
print("-----------------------------------------------------")
print(f"Non-Fraudulent (Class 0): {train_counts[0]} samples ({train_counts[0]/total_train*100:.4f}%)")
print(f"Fraudulent     (Class 1): {train_counts[1]} samples ({train_counts[1]/total_train*100:.4f}%)")
print("=====================================================")
```

```python
import os
import numpy as np

# Use the same base notebook directory
notebook_dir = os.getcwd()

# 1. Define paths for the resampled NumPy files
x_train_path = os.path.abspath(os.path.join(notebook_dir, "..", "..", "data", "x_train_resampled.npy"))
y_train_path = os.path.abspath(os.path.join(notebook_dir, "..", "..", "data", "y_train_resampled.npy"))

# 2. Load the numpy arrays
x_train_resampled = np.load(x_train_path)
y_train_resampled = np.load(y_train_path)

# 3. Print the shapes and counts
print("=============================")
print(f"X_train resampled shape: {x_train_resampled.shape}")
print(f"Y_train resampled shape: {y_train_resampled.shape}")
print("=============================")
print(f"Total Resampled Samples: {len(y_train_resampled)}")

# Quick check on the class balance (since it's a 1D target array)
unique_classes, counts = np.unique(y_train_resampled, return_counts=True)
for cls, count in zip(unique_classes, counts):
    print(f"Class {cls} count: {count}")
print("=============================")
```
