```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("../data/creditcard.csv")
x= dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
print(x)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42,stratify=y)
```

```python
from sklearn.preprocessing import RobustScaler
rob_scaler = RobustScaler()
x_train[:,[0,29]] = rob_scaler.fit_transform(x_train[:,[0,29]])
x_test[:,[0,29]]  = rob_scaler.transform(x_test[:,[0,29]])
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
print(f"Train Fraud Ratio: {np.sum(y_train) / len(y_train):.4f}")
print(f"Test Fraud Ratio: {np.sum(y_test) / len(y_test):.4f}")
```

```python
# Performing UnderSampling on the dataset
train_df = pd.DataFrame(x_train)
train_df['class'] = y_train

fraud_indices = train_df[train_df['class'] == 1].index
non_fraud_indices =  train_df[train_df['class'] == 0].index

random_non_fraud_indices = np.random.choice(non_fraud_indices, len(fraud_indices), replace=False)

under_sample_indices = np.concatenate([fraud_indices,random_non_fraud_indices])
under_sample_train = train_df.loc[under_sample_indices]

print(f"New Dataset Shape: {under_sample_train.shape}")
print("New Class Distribution:")
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=42)

x_train_resampled, y_train_resampled = sm.fit_resample(x_train, y_train)

print(f"New Dataset Shape: {x_train_resampled.shape}")
print("New Class Distribution (y_train_resampled):")
print(pd.Series(y_train_resampled).value_counts())
```

```python
#saving resampled training data and original test data
np.save('../data/x_train_resampled.npy', x_train_resampled)
np.save('../data/y_train_resampled.npy', y_train_resampled)

# Save the original test data (we didn't SMOTE this, remember?)
np.save('../data/x_test.npy', x_test)
np.save('../data/y_test.npy', y_test)

print("Baton passed! Data saved to the data folder")
```

```python
# Saving the scaler using pickle
import pickle
with open("../models/scaler.pkl", "wb") as file:
    pickle.dump(rob_scaler,file)
print("RobustScaler saved sucessfully")
```
