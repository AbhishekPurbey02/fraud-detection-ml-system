```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import sklearn
```

```python
# Loaded the resampled data
x_train = np.load('../data/x_train_resampled.npy')
y_train = np.load('../data/y_train_resampled.npy')
x_test = np.load('../data/x_test.npy')
y_test = np.load('../data/y_test.npy')

print("Ready to train! Resampled dataset loaded")
```

```python
# Implemented the Logistic Regression and predicted and checked the results on it 
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report

lr_model = LogisticRegression(max_iter=1000)
lr_model.fit(x_train,y_train)
y_pred = lr_model.predict(x_test)

print("=== Confusion Matrix ===")
print(confusion_matrix(y_test,y_pred))

print("=== Classification Report ===")
print(classification_report(y_test,y_pred))

```

```python
# Fine Tuning the logistics regression algrothim model by playing with the threshold
from sklearn.metrics import recall_score, precision_score

y_probs = lr_model.predict_proba(x_test)[:,1]
thresholds = [0.3,0.5,0.7,0.9]

print(f"{'Threshold':<12} | {'Precision':<10} | {'Recall':<10}")
print("-"*35)

for t in thresholds:
    y_pred_t = (y_probs>= t).astype(int)

    p = precision_score(y_test,y_pred_t)
    r = recall_score(y_test,y_pred_t)

    print(f"{t:<12} | {p:<10.4f} | {r:<10.4f}")
```

```python
# Implemented Random Forest Classifier and evaluated the results
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report

rf_model = RandomForestClassifier(n_estimators=50,random_state=42,n_jobs=-1) 

rf_model.fit(x_train,y_train)
y_pred_rf = rf_model.predict(x_test)

print("=== Confusion Matrix")
print(confusion_matrix(y_test,y_pred_rf))

print("===Classification Report===")
print(classification_report(y_test,y_pred_rf))
```

```python
# Fine tuning Random Forest by changing prediction threshold

y_probs_rf = rf_model.predict_proba(x_test)[:,1]

thresholds = [0.3,0.5,0.7,0.9]
for thresholds in thresholds:
    y_pred_threshold_rf = (y_probs_rf>=thresholds).astype(int)

    print(f"\nThreshold: {thresholds}")
    print(f"Recall:", recall_score(y_test,y_pred_threshold_rf))
    print(f"Precision:", precision_score(y_test,y_pred_threshold_rf))
    print(f"confusion_matrix:")
    print(confusion_matrix(y_test,y_pred_threshold_rf))
```
